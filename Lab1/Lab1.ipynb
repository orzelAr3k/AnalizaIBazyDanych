```python
# import
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
# funkcja
f = lambda x: x ** 2 + 5
```

```python
x1 = np.arange(-1, 1, 0.1)
plt.plot(x1, f(x1), label="Funkcja $x^2+5$")
plt.title("Wykres fukcji $x^2+5$ na przedziale $-1<x<1$")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
```

```python
x2 = np.arange(-6, 6, 0.1)
plt.plot(x2, f(x2), label="Funkcja $x^2+5$")
plt.title("Wykres fukcji $x^2+5$ na przedziale $-6<x<6$")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
```

```python
x3 = np.arange(0, 5, 0.1)
plt.plot(x3, f(x3), label="Funkcja $x^2+5$")
plt.title("Wykres fukcji $x^2+5$ na przedziale $-0<x<5$")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
```
